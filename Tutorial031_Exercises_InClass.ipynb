## Part I: Start with some data.
* First plot it
* Then compute summary stats (mean, std) for each data array
* Then compute the correlation coeffecient that relates the two arrays
* Then the t-value and p-value associated with the correlation.

[source of this famous data set: Anscombe](https://en.wikipedia.org/wiki/Anscombe%27s_quartet)

```python
d1 = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
d2 = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])
```

## Plot data...scatter works well here
* First rule of data analysis...always plot your data first!

```python
plt.scatter(d1, d2, color='b')
plt.xlabel('D1 Data')
plt.ylabel('D2 Data')
plt.show()
```

## Compute mean, std of both data arrays
* Leave the output in the notebook so that we can come back and compare later

```python
d1_mean = d1.mean()
d2_mean = d2.mean()
d1_std = d1.std()
d2_std = d2.std()

print ('D1 Mean is', d1_mean, 'and STD is', d1_std,'\n', 'D2 Mean is', d2_mean, 'and STD is', d2_std)
```

## Correlate the two data arrays, compute t-value and p-value associated with correlation coef

```python
# Correlation coefficient
corr_coe = np.corrcoef(d1,d2)[0,1]
print('Correlation coefficient:', corr_coe)

# Computing t-value using r 
# t = r * sqrt[(n-2)/(1-r^2)]
t_value = (corr_coe * np.sqrt (len(d1)-2) / np.sqrt(1 - corr_coe**2))
print('T-value:', t_value)

#Computing p-value using r
df = len(d1)-1                 # degrees of freedom
p_val = (1-stats.t.cdf(t_value, df))*2
print('P-value:', p_val)
```

## Now use randomization testing to eval the reliability of the estimated p-value. 

<div class="alert alert-info">
Before you run this procedure, you should have a pretty good idea about how the standard p-value will compare with the p-value that you estimate using randomization. Make a prediction!
</div>

```python
# Parametric makes a lot of assumptions about that data that are 
# probably untrue, so the two p-values will not be very similar

# Randomization wil produce a more accurate p-value (which is probably larger)

num_randoms = 1000        # number of times to run randomizations
corr_randoms = np.zeros(num_randoms)   # array to hold correlation coefficients of each run
t_val_rand = np.zeros(num_randoms)     # emtpy to hold t-values of each run
tmp0 = np.zeros(len(d1))        # array, length of d1
tmp1 = np.zeros(len(d2))        # array, lenght of d2

# as with in class tutorial030
# Loop to iterate over randomization trials (1000)
for i in np.arange(num_randoms):
    
    #switch whether data coming from d1 or d2 (condition doesn't matter)
    # so randomly choose between 0 and 1 assign to either tmp0 or tmp1
    for j in np.arange(len(d1)):
        if np.random.randint(2) == 0:
            tmp0[j] = d1[j]
            tmp1[j] = d2[j]
        else:
            tmp0[j] = d2[j]
            tmp1[j] = d1[j]
            
    # correlate the two randomized sets (tmp0 and tmp1)
    corr_randoms[i] = np.corrcoef(tmp0,tmp1)[0,1]
    t_val_rand[i] = corr_randoms[i] * np.sqrt(len(d1)-2) / np.sqrt(1 - corr_randoms[i]**2)

# Compute p-value of randomization trials
p_val_rand = 2*(1-(np.sum(t_value>t_val_rand) / num_randoms))

print('Randomization-based p-value:', p_val_rand, '\n','Parametric p-value:', p_val)


```

## Part III: Load in the second EEG data set from last week (eeg_data01.npz). 
* Pull out the data, sr, and tx arrays

```python
# load the data...
eeg = np.load('eeg_data01.npz')

# Pull out data, sr, and tx arrays
data = eeg['data']
sr = eeg['sr']
tx = eeg ['tx']

data.shape
```

## Below is a vector that labels each trial as coming from experimental conditions 1,2,3,4

```python
N = data.shape[0]
num_samp_cond = int(N/4)
cond = np.hstack((np.ones(num_samp_cond), np.ones(num_samp_cond)*2, np.ones(num_samp_cond)*3, np.ones(num_samp_cond)*4))
```

## Now avgerage the data from condition 2, avg the data from condition 3, and plot against the time axis (tx) - we're going to ignore conditions 1 and 4

```python
cond_2 = data[cond==2,:]
cond_3 = data[cond==3,:]

cond_2_avg = np.mean(cond_2, axis=0)
cond_3_avg = np.mean(cond_3, axis=0)

plt.plot(tx, cond_2_avg, color='r', linewidth = 1.5)
plt.plot(tx, cond_3_avg, color='b', linewidth = 1.5)
plt.xlabel('Time')
plt.ylabel('Mean')
plt.show()
```

## Next do a t-test for related samples comparing the responses in conditions 2 and 3 at each point in time. Note - you can do this all in one line of code!

```python
t_values_2_3 = stats.ttest_ind(cond_2, cond_3, axis=0)
```

## now plot the averages in one plot, and then right below that make another plot with the t-values. Keep in mind that with this many degrees of freedom, a t-value of approx 1.9 is significant at the magic 0.05 level (or 1.68 one-tailed). So put some horizontal lines on the plot at 1.9 and -1.9. You'll see some pretty impressive t-values in the data!

```python
plt.figure(1)
plt.plot(tx, cond_2_avg, color='g', linewidth = 1.5)
plt.plot(tx, cond_3_avg, color='b', linewidth = 1.5)
plt.legend(['Condition 2', 'Condition 3'])
plt.xlabel('Time')
plt.ylabel('Mean')

plt.figure(2)
plt.plot(tx, t_values_2_3.statistic, color='b', linewidth = 1.5)
plt.axhline(1.9, color='r', alpha=0.5)
plt.axhline(-1.9, color='r', alpha=0.5)
plt.xlabel('Time')
plt.ylabel('T-value')

plt.show()
```

## You should see that there is a significant difference at many timepoints...Now figure out how robust those effects are by applying bootstrapping 
* To do this, you might first create two 800 x 4102 matrices, one with data from all trials of condition 2, and one with data from all trials of condition 3
* Then resample 800 trials, with replacement, from each data matrix and then do the t-test. 
* try generating a set of 800 values with repeating numbers that you can use for a row index into the data matrices
* repeat and then compute CIs of the t-value
* how often do the CIs for the t-value overlap with 0???
* note - this can take a while, so start with 50 bootstraps and then increase as compute time allows

```python
# Matrices created above as cond_2 and cond_3

# storing the number of trials
N = cond_2.shape[0]
print(N)

# Number of bootstraps and an array of zeros with dimensions of:
# number of bootstraps x time points
num_bootstraps = 800
boot_t_values = np.zeros((num_bootstraps,data.shape[1]))

# Loop that will sample with replacement and then do the t-test
# and store the value

for i in np.arange(num_bootstraps):
    # With replacement generate a sample number from 0:N 
    # exclusive and do that N times
    index2 = np.random.randint(0, high=N, size=N)
    index3 = np.random.randint(0, high=N, size=N)
  
    # Pull data using index
    a = cond_2[index2]
    b = cond_3[index3]
    
    #compute T-test
    t_val_ab = stats.ttest_ind(a, b, axis=0)
  
    #store t-values
    boot_t_values[i,:] = t_val_ab.statistic

# then compute 95% CIs of T-values
CIs = np.percentile(boot_t_values, [2.5, 97.5], axis=0)

plt.plot(tx, np.mean(boot_t_values, axis=0), color='b', linewidth = 1.5)
plt.plot(tx, CIs[0,], color='r', linewidth = 1.0)
plt.plot(tx, CIs[1,], color='r', linewidth = 1.0)
plt.axhline(0, color='k', linewidth=0.5)
plt.xlabel('Time')
plt.ylabel('T-value')

plt.show()
```
