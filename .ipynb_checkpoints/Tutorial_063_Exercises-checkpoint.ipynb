## First import data from eeg_data00.npz

```python
# first import the modules that you'll need. 
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt

# import our sinc filter implementation function...
import my_sinc_filt as msf

# also define the default font we'll use for figures. 
fig_font = {'fontname':'Arial', 'size':'20'}
```

## Just like in the first FFT tutorial load eeg_data00.npz so that we have an EEG data set to play with
* you should find that it has three arrays: data, tx, and sr (with 'data' a #trials x # timepoints array containing the EEG data, the time of each data acquisition in tx, and the sample rate stored as sr).  

```python
# load the data...
eeg = np.load('eeg_data00.npz')

# get the different arrays like this...kind of like a structure in matlab. 
eeg['data']
eeg['sr']
eeg['tx']

# and can query the attributes of the data like this...which will tell us that there 960 trials and 876 timepoints per trial sampled
# at 250Hz
print('Shape of the big eeg data set: ', eeg['data'].shape)
print('Sample rate: ', eeg['sr'])

# and if you want to save some typing, especially because we only have a few variables, you reassign the different arrays like this
data = eeg['data']
sr = eeg['sr']
tx = eeg['tx']
# now you can just write data instead of eeg['data'] (and can for tx and sr as well...)
```

## Extract the data from time 0 until time 2000
* hint: you can use boolean indexing by selecting the columns from the 'data' matrix that meet: (tx>=0)  & (tx<2000)
* while you're at it, go ahead and grab the timepoints corresponding to that interval as well

```python
# Boolean indexing to grab data from timepoints 0 to 2000
data_interval = data[:,(tx>=0) & (tx<2000)]
time_interval = tx[(tx>=0) & (tx<2000)]

# Plot to mean over the 0 to 2000 interval to make sure 
# interval correct
mean_interval = np.mean(data_interval, axis = 0)

plt.plot(time_interval, mean_interval)
plt.xlabel('Time Interval 0-2000', **fig_font)
plt.ylabel('Mean EEG', **fig_font)
plt.show()
```

## Then FFT each trial and take a look at the average power across frequencies
* FFT each trial 
* Then compute the amplitude on each trial
* Then compute the average amplitude across all trials

```python

# Store shape of data matrix 
r, c = data_interval.shape

# FFT each trial (axis is 1, so indexing by row)
fft_trials = np.fft.rfft(data_interval, axis = 1)

# Computing amplitute for each trial and then averaging
# amplitude across all trials for a specific timepoint 
# (indexing by column, so one specific timepoint)
amplitude = (np.abs(fft_trials) / c ) * 2
avg_amp = np.mean(amplitude, axis = 0)

# Create frequency axis (1/sr = dt = time step)
# rfftfreq --> return the Discrete Fourier Transform sample frequencies
freq_axis = np.fft.rfftfreq(c, 1/sr)

# Plot
plt.plot(freq_axis, avg_amp)
plt.title('Normalized EEG FFT')
plt.xlabel('Frequency', **fig_font)
plt.ylabel('Amplitude', **fig_font)
plt.show()
```

## Notice that there are two prominent features in the freq domain representation. Make a lp filter that gets rid of power > 20Hz (you'll know if it worked if you get rid of the spike at 60Hz)
* Because our apply_filter function isn't optimized to work on matrices (we could do that, but haven't) you will need to loop and apply filter to each trial

```python
fc = 0.1
b = 0.1
win = 'blackman'
type_filt = 'lp'

# make our function call to generate a lp filter
n, sinc_func = msf.define_filt(fc=fc, b=b, window=win, type_filt=type_filt)

# filter the signal (as in class)
filt_sig = np.zeros((r,c))
for i in np.arange(r):
    filt_sig[i,:] = msf.apply_filt(input_sig=data_interval[i,:], input_filter=sinc_func)
```

## Then FFT it again to see if you got rid of the hf components of the signal

```python
# Store shape of data matrix 
r, c = filt_sig.shape

# FFT each trial (axis is 1, so indexing by row)
fft_lp = np.fft.rfft(filt_sig, axis = 1)

# Computing amplitute for each trial and then averaging
# amplitude across all trials for a specific timepoint 
# (indexing by column, so one specific timepoint)
amp_lp = (np.abs(fft_lp) / c ) * 2
avg_amp_lp = np.mean(amp_lp, axis = 0)

# Create frequency axis (1/sr = dt = time step)
# rfftfreq --> return the Discrete Fourier Transform sample frequencies
freq_axis_lp = np.fft.rfftfreq(c, 1/sr)

# Plot
plt.plot(freq_axis_lp, avg_amp_lp)
plt.title('Normalized EEG FFT With Low Pass Filter at 20 Hz')
plt.xlabel('Frequency', **fig_font)
plt.ylabel('Amplitude', **fig_font)
plt.show()
```

## Now design a band pass filter that gets rid of everything outside of the 6-14Hz range (where that big bump is...)
* To design a band pass filter, make a lp filter (with an appropriate fc), then make a hp filter (with an appropriate fc)
* use np.convolve to combine the two filters 
* then call the apply_filt function
* Then you should be left with only those frequencies that pass the lp and the hp filters (i.e. a band pass filter)!

```python
# Make low pass filter below 14 Hz
fc_lp = 14 / sr
# Make a high pass filter above 6 Hz
fc_hp = 6 / sr

b = 0.1
win = 'blackman'
type_filt = 'lp'

# make our function call to generate a lp filter
n, sinc_func_lp = msf.define_filt(fc=fc_lp, b=b, window=win, type_filt=type_filt)

type_filt = 'hp'

# make our function call to generate a hp filter
n, sinc_func_hp = msf.define_filt(fc=fc_hp, b=b, window=win, type_filt=type_filt)

# convolve two filters
b_filt = np.convolve(sinc_func_lp,sinc_func_hp)
plt.plot(b_filt)

# Filter signal
filtered_signal = np.zeros((r,c))
for i in np.arange(r):
    filtered_signal[i,:] = msf.apply_filt(input_sig=data_interval[i,:], input_filter=b_filt)
```

```python
# Store shape of data matrix 
r, c = filtered_signal.shape

# FFT each trial (axis is 1, so indexing by row)
fft_band = np.fft.rfft(filtered_signal, axis = 1)

# Computing amplitute for each trial and then averaging
# amplitude across all trials for a specific timepoint 
# (indexing by column, so one specific timepoint)
amp_band = (np.abs(fft_band) / c ) * 2
avg_amp_band = np.mean(amp_band, axis = 0)

# Create frequency axis (1/sr = dt = time step)
# rfftfreq --> return the Discrete Fourier Transform sample frequencies
freq_axis_band = np.fft.rfftfreq(c, 1/sr)

# Plot
plt.plot(freq_axis_band, avg_amp_band)
plt.title('Normalized EEG FFT With Band Filter 6-14 Hz')
plt.xlabel('Frequency', **fig_font)
plt.ylabel('Amplitude', **fig_font)
plt.show()
```

## You can also make a band-stop filter using a similar logic
* Instead of convolving the lp and the hp filters, you can add them together
* Use this to selectively get rid of the 60Hz line noise in the data by generated a band-stop filter over 50-70Hz

```python
# Make low pass filter below 14 Hz
fc_lp = 50 / sr
# Make a high pass filter above 6 Hz
fc_hp = 70 / sr

b = 0.1
win = 'blackman'
type_filt = 'lp'

# make our function call to generate a lp filter
n, sinc_func_lp = msf.define_filt(fc=fc_lp, b=b, window=win, type_filt=type_filt)

type_filt = 'hp'

# make our function call to generate a hp filter
n, sinc_func_hp = msf.define_filt(fc=fc_hp, b=b, window=win, type_filt=type_filt)

# convolve two filters
b_stop_filt = sinc_func_lp + sinc_func_hp
plt.plot(b_stop_filt)

# Filter signal
filtered_signal = np.zeros((r,c))
for i in np.arange(r):
    filtered_signal[i,:] = msf.apply_filt(input_sig=data_interval[i,:], input_filter=b_stop_filt)
```

## Then FFT and look at filtered signal in the frequency domain

```python
# Store shape of data matrix 
r, c = filtered_signal.shape

# FFT each trial (axis is 1, so indexing by row)
fft_band = np.fft.rfft(filtered_signal, axis = 1)

# Computing amplitute for each trial and then averaging
# amplitude across all trials for a specific timepoint 
# (indexing by column, so one specific timepoint)
amp_band = (np.abs(fft_band) / c ) * 2
avg_amp_band = np.mean(amp_band, axis = 0)

# Create frequency axis (1/sr = dt = time step)
# rfftfreq --> return the Discrete Fourier Transform sample frequencies
freq_axis_band = np.fft.rfftfreq(c, 1/sr)

# Plot
plt.plot(freq_axis_band, avg_amp_band)
plt.title('Normalized EEG FFT With Band Stop Over 50-70 Hz')
plt.xlabel('Frequency', **fig_font)
plt.ylabel('Amplitude', **fig_font)
plt.show()
```
