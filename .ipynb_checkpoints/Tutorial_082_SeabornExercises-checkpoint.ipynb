# Seaborn Exercises

1) From the "tips" dataset
    1. add a percentage tip column to the dataset
    2. visualize the relationship between tip percentage, with day of week and time 
    3. visualize tip ~ day, time, and number of people (which function allows three variables independent variables at once?)


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame

# Load tips dataset into dataframe
tips = sns.load_dataset('tips')

# Add percentage tip column to the dataset
tips['tip_percentage'] = tips.tip / tips.total_bill
tips.head()
```

```python
# Visualize relationship between tip percentage 
# with day of week and time 

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(x = 'time', y = 'tip_percentage', hue= 'day', data=tips)
plt.subplot(1,2,2)
sns.boxplot(x = 'time', y = 'tip', hue= 'day', data=tips)
plt.show()
```

```python
# Visualize relationship between tip percentage 
# with day of week and time 

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.violinplot(x = 'time', y = 'tip_percentage', hue= 'day', data=tips)
plt.subplot(1,2,2)
sns.violinplot(x = 'time', y = 'tip', hue= 'day', data=tips)
plt.show()
```

```python
sns.pointplot(x = 'day', y = 'tip_percentage', hue ='time', data = tips)
```

```python
# visualize tip ~ day, time, and number of people 
# (which function allows three variables independent 
# variables at once?)

sns.factorplot(x='size',y='tip_percentage', hue='time',col='day',data=tips,kind='bar')
plt.show()
```

2) Find an interesting trend in the "exercise" dataset

```python
exercise = sns.load_dataset('exercise')
del exercise ['Unnamed: 0']
exercise.head()
```

```python
sns.factorplot(x='time', y = 'pulse', hue = 'diet', col ='kind', data = exercise, kind = 'point')
plt.show()

# Demonstrates that those on low fat diets have lower pulses during 
# prolonged exercise compared to no fat diet
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.barplot(x = 'kind', y = 'pulse', data = exercise, hue = 'diet')
plt.subplot(1,2,2)
sns.pointplot(x = 'kind', y = 'pulse', data = exercise, hue = 'diet')
plt.show()
```
