## Read in data from eeg study
* This time I've kept in data across a 1s window from many electrodes (20) sampled at 250Hz, so now we have a [960, 20, 250] matrix that contains data from all 960 trials, 20 electrodes, and over a 1s window following stimulus onset (250 timepoints)
* Also read in a "cond_labels" vector that has 960 0's and 1's to mark trials from the two experimental conditions 

## import stuff...

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh

# Support vector classifier
from sklearn.svm import SVC 

fig_font = {'fontname':'Arial', 'size':'20'}
```

```python
# load the data...
eeg = np.load('/Users/IMC/Documents/GIT_HELPER/class_demo_eeg.npz')

# get the different arrays like this...
eeg['data']
eeg['sr']
eeg['tx']
eeg['cond_labels']

# shape of data
print('Shape of the big eeg data set: ', eeg['data'].shape)

# and if you want to save some typing, especially because we only have a few variables, 
# you reassign the different arrays like this
# Condition labels 0 and 1 for the two different conditions
data = eeg['data']
sr = eeg['sr']
tx = eeg['tx']
cond_labels = eeg['cond_labels']
```

## Compute the mean across the first dimension (trials) and then plot the mean response across time in each of the 20 electrodes on one axis

* Might need to transpose the data into a Time x Electrode format

```python
# [960, 20, 250] matrix --> 960 trials, 20 electrodes, 250 timepoints

# Computing mean across trials, which makes [20,250] array 
# representing [Electrode x Time]
trial_mean = np.mean(data, axis = 0)

# Transpose data into Time x Electrode format for plot
plt.plot(tx, trial_mean.T)
plt.xlabel('Time (ms)')
plt.ylabel('EEG Mean')

plt.show()

```

## That gives you an idea about what the pattern of data looks like across electrodes. 
* Next, take the mean across time, so you have a 960 x 20 matrix 
* Use the cond_labels vector to sort the data based on condition (so you have two 480 x 20 matrices)
* Do classification to see if you can decode the experimental condition
* Remember - break up your data into independent training and testing sets
* Set C to a small number if this is taking a long time...

```python
# [960, 20, 250] matrix --> 960 trials, 20 electrodes, 250 timepoints

mean_across_time = np.mean(data, axis = 2)
print (mean_across_time.shape)
print (cond_labels.shape)

cond_1 = np.empty([480, 20])
cond_2 = np.empty([480, 20])

x = 0 
for i in cond_labels:
    if i == 0:
        np.append(cond_1, mean_across_time[x])
        x = x + 1
    if i == 1:
        np.append(cond_2, mean_across_time[x])
        x = x + 1
        
print (cond_1.shape)
print (cond_2.shape)

# Other way of doing this from Chris & Meghan

cond_1b = mean_across_time[np.where(cond_labels == 0)]
cond_2b = mean_across_time[np.where(cond_labels == 1)]

print (cond_1b.shape)
print (cond_2b.shape)

# Do classification to see if you can decode the experimental condition
# Remember - break up your data into independent training and testing sets
# Set C to a small number if this is taking a long time...

# 90% of data for training, 10% for test
r, c = cond_1.shape    # shape of the array
trn_length = int(np.round(.9*r,0)) 
tst_length = r-trn_length

# Stacking the first 90% from condition 1 and 2 into training data
# Stacking the last 10% from condition 1 and 2 into test data
trn_data = np.vstack((cond_1[0:trn_length,:], cond_2[0:trn_length,:]))
tst_data = np.vstack((cond_1[trn_length,:], cond_2[trn_length,:]))

# Vectors to hold condition 1 or condition 2 label (horizontally
# stack so get [864, ] array )
trn_labels = np.hstack((np.zeros(trn_length), np.ones(trn_length)))
tst_labels = np.hstack((np.ones(tst_length), np.ones(tst_length)))

# Train the classifer using training data using linear SVM 
# Starting with small C

model = SVC(kernel='linear', C=10)

# Fit model to training data (passing in data matrix and experimental
# condition label for each trial)
model.fit(trn_data, trn_labels)
```
